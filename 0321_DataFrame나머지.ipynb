```python
# DataFrame 연결 (행방향)

import numpy as np
import pandas as pd

df1 = pd.DataFrame({'a':['a0','a1','a2','a3'],
                    'b' : [1,2,3,4],
                    'c' : ['c0','c1','c2','c3']},
                    index=[0,1,2,3])

df2 = pd.DataFrame({ 'b' : [5,6,7,8],
                  'c' : ['c0','c1','c2','c3'],
                  'd' : ['d0','d1','d2','d3'],
                  'e' : ['e0','e1','e2','e3']},
                  index=[2,3,4,5])
display(df1)
display(df2)

# 기존으 index를 유지하면서 연걸
result_df = pd.concat([df1,df2],axis=0)
display(result_df)

# 기존의 index를 없애고 숫자 index를 다시 설정
result_df = pd.concat([df1,df2], axis=0,
                       ignore_index=True)
display(result_df)
```

```python
# DataFrame의 결합(merge)
data1 = {'학번' : [1,2,3,4],
          '이름' : ['홍길동','김연아','아이유','신사임당'],
          '학년':[1,2,3,4]}

df1 = pd.DataFrame(data1)
display(df1)

data2 = {'학번' : [1,2,3,4],
         '학과' : ['철학과','수학과','컴퓨터','국어국문'],
         '학점' : [1.2, 3.3, 2.7, 4.0]}
df2 = pd.DataFrame(data2)
display(df2)

result_df1 = pd.merge(df1,df2, on='학번', how='inner')   # inner join
display(result_df1)
```

```python
# DataFrame의 결합(merge)
data1 = {'학번' : [1,2,3,4],
          '이름' : ['홍길동','김연아','아이유','신사임당'],
          '학년':[3,2,1,4]}

df1 = pd.DataFrame(data1)
display(df1)

data2 = {'학번' : [1,2,4,5],
         '학과' : ['철학과','수학과','컴퓨터','국어국문'],
         '학점' : [1.2, 3.3, 2.7, 4.0]}

df2 = pd.DataFrame(data2)
display(df2)


result_df = pd.merge(df1, df2, on='학번', how='outer')  # outer join
display(result_df)

result_df = pd.merge(df1,df2, on='학번', how='left')   # left outer join
display(result_df)

result_df = pd.merge(df1,df2, on='학번', how='right')   # right outer join
display(result_df)
```

```python
# DataFrame의 결합(merge)
# outer join
# inner join에 포함되지 않는 놈들도 포함.

data1 = {'학번' : [1,2,3,4],
          '이름' : ['홍길동','김연아','아이유','신사임당'],
          '학년':[3, 2, 1, 4]}

df1 = pd.DataFrame(data1)
display(df1)

data2 = {'학생학번' : [1, 2, 4, 5],
         '학과' : ['철학과','수학과','컴퓨터','국어국문'],
         '학점' : [1.2, 3.3, 2.7, 4.0]}

df2 = pd.DataFrame(data2)
display(df2)


# 결합하려는 기준 column명이 서로 다른 경우
result_df = pd.merge(df1, df2, 
                     left_on='학번', 
                     right_on='학생학번', 
                     how='inner') 
display(result_df)
```
