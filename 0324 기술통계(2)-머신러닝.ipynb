```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./data/student_scores_em.csv',
                 index_col='student number')
# display(df.head())
# print(df.shape)

en_scores = df['english'][:10]
# print(en_scores)  # series
ma_scores = df['mathematics'][:10]

scores_df = pd.DataFrame({'english' : en_scores.values,
                          'mathmatics' : ma_scores.values},
                          index=['A','B','C','D','E','F','G','H','I','J'])

scores_df['en_deviation'] = scores_df['english'] - scores_df['english'].mean()
scores_df['ma_deviation'] = scores_df['mathmatics'] - scores_df['mathmatics'].mean()
scores_df['product_deviation'] = scores_df['en_deviation'] * scores_df['ma_deviation']

display(scores_df)
```

```python
df = pd.read_csv('./data/student_scores_em.csv',
                 index_col='student number')
# display(df.head())
# print(df.shape)

en_scores = df['english'][:10]
# print(en_scores)  # series
ma_scores = df['mathematics'][:10]

scores_df = pd.DataFrame({'english' : en_scores.values,
                          'mathmatics' : ma_scores.values},
                          index=['A','B','C','D','E','F','G','H','I','J'])

# 공분산 -> numpy함수 이용
# 결과가 covariance mattrix(공분산 행렬)로 리턴
# 편차의 곱의 평균을 구하는데 평균을 구할 때 n으로 나누는 경우와 n-1로 나눈는 경우
#   영어,영어(영어의 공분산) / 영어,수학
#           수학,영어       / 수학,수학(수학의 공분산)

cov_matrix = np.cov(en_scores, ma_scores,ddof=0)
print(cov_matrix)

# 참고로 pandas에도 cov() 존재! 그런데 pandas의 cov()는 무조건
# n-1로 나누는 불편분산상태로 사용됨!
```

```python
# 실제 데이터를 이용해서 공분산을 구해보자
# KOSPI 지수와 삼성전자 주가에 대한 공분산을 구해보자
# 2018년도 기준으로 KOSPI 200안에서 삼성전자 차지하는 비율은 33%

# Yahoo finance에서 주가데이터를 받아서 사용함
import pandas_datareader.data as pdr # 설치해야함
# pip install pandas_datareader
from datetime import datetime

start = datetime(2018,1,1)  # 특정날짜에 대한 날짜객체를 생성
end = datetime(2018,12,31)
print(start)

df_kospi = pdr.DataReader('^KS11','yahoo',start,end)
display(df_kospi.head())
df_se = pdr.DataReader('005930.KS','yahoo',start,end)
display(df_se)

# 공분산
cov_matrix = np.cov(df_kospi['Close'], df_se['Close'],ddof=0)
print(cov_matrix)   # [[   24078.14439225   488212.99831291]
                    # [  488212.99831291 11871059.41111261]]
                    # -> 양의 상관관계
    
# 다른 종목으로 공분산 구하기 
# 남북경험주, 방산주를 예로 들어서 음의 상관관계에 대해 알아보자
# 남북경험주 => 부산산업(011390.KS)
# 방산주 => LIG넥스원(079550.KS)
df_busan = pdr.DataReader('011390.KS','yahoo',start,end)
df_lig = pdr.DataReader('079550.KS','yahoo',start,end)

cov_matrix = np.cov(df_busan['Close'], df_lig['Close'],ddof=0)
print(cov_matrix)   # -3.84951773e+08  # 음의 상관관계

```
