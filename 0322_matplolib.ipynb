```python
# matplotlib에 대해 알아봐요

# 1. Lineplot (선 그래프)

import matplotlib.pyplot as plt

# plot 제목 설정
plt.title('Line Plot')

# plot()이라는 메소드를 이용해서 line plot을 그릴 수 있음
# x 축의 자료 위치 (x축 눈금) -> tick은 기본적으로 0, 1, 2, 3
# plt.plot([1, 5, 12, 25])
# 만약 tick을 별도록 명시하고 싶으면 아래와 같이 하면 됨
plt.plot([10,20,30,40],[1,2,3,4])


# show() 함수 호출 -> 화면에 rendering하고 마우스 이벤트를 기다리는 함수
# 만약 코드를 pycharm과 같은 일반 개발 톨을 이용하면 이 명령어를 이용해야 그래프가 나옴
# 하지만 jupyter notebook은 셀 단위로 plot을 rendering하기 때문에 굳이 show()를 할 필요 없음
plt.show()
```

```python
# 간단한 데이터를 이용해서 line plot의 여러가지 기능들을 알아봐요
# 데이터는 제공된 excel 이용
# pandas를 이용해서 excel 파일을 DataFrame으로 변환

import numpy as np
import pandas as pd
import warnings

df = pd.read_excel('./data/lineplot_sample_data.xlsx')

display(df.head())

# NaN을 채우기위해 ffill을 이용. 이전 행의 값으로 NaN을 채움
df = df.fillna(method = 'ffill')

# display(df)

# 관심있는 데이터 추출
# 서울특별시에서 다른 지역으로 이동한 데이터만 추출  ->  boolean indexing 이용
# boolean indexing을 이용해서 원하는 데이터 추출
# boolean mask를 만들어서 indexing 하기

boolean_mask = (df['전출지별'] == '서울특별시')  & (df['전입지별'] != '서울특별시')
df_seoul = df.loc[boolean_mask,:]
# display(df_seoul)

df_seoul.drop('전출지별', axis=1, inplace=True)
# display(df_seoul)

# '전입지별' -> '전입지'로 column명을 변경
df_seoul.rename({'전입지별' : '전입지'}, axis=1, inplace=True)
# display(df_seoul)

# '전입지' column을 행 index로 설정
df_seoul.set_index('전입지', inplace=True)
display(df_seoul)

# 서울특별시에서 경기도로 전입한 데이터만 가져오기
s = df_seoul.loc['경기도']

print(s)

plt.plot(s.index, s.values)
plt.title('seoul to gyeongido')
plt.show()
```

```python
# 한글 깨지는 현상 해결
# 한글 폰트가 있어야함 -> 구글 드라이브에 공유
# 필요한 module import

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib import rc

import warnings
import numpy as np
import pandas as pd

warnings.filterwarnings(action='ignore')   # 경고메시지 출력되지 않게 설정

# 그래프에서 '-' 기호때문에 문제가 발생할 여지가 있음
mpl.rcParams['axes.unicode_minus'] = False

# 한글폰트 사용방법은 여러가지 존재
font_path = './font/malgun.ttf'
font_name = fm.FontProperties(fname=font_path).get_name()
print(font_name)   # Malgun Gothic
rc('font',family=font_name)

# 이전 셀에서 만들어둔 serires(연도별 경기도롤 전입한 인구수)를 이용해서
# line plot 그림
plt.plot(s.index, s.values)

plt.title('서울에서 경기도로 전입한 인구 추이')
plt.xlabel('연도')
plt.ylabel('이동 인구수')
plt.show()
```

```python
# 약간의 style 적용
# 그래프에 스타일 적용
plt.style.use('ggplot')

# 그래프 크기 설정
plt.figure(figsize=(10, 5))   # 단위는 인치, 가로, 세로순으로 크기 설정

# x의 눈금을 읽기 어려움 ->  label을 회전시킴
plt.xticks(rotation='vertical')

plt.plot(s.index, s.values,
         marker = 'o',  # o는 circle marker, s는 square marker
         markersize=8,
         markerfacecolor='b',
         color='g',   # 선의 색깔
         linewidth=2)
plt.title('서울에서 경기도로 전입한 인구 추이')
plt.xlabel('연도')
plt.ylabel('이동 인구수')

plt.legend(labels=['서울 -> 경기'],loc='best')
plt.show()
```

```python
# sub_plot에 대해 알아보자
# => 쉽게 말하면 그래프 여러개를 한번에 그리는 방법

# Figure, Axes, Axis에 대해 간단히 알아보자
# Figure는 canvas를 나타내는 객체, 쉽게 말하면 도화지 역할을 함

# jupyter notebook은 plot()을 사용하려면 자동으로 figure객체를 하나 만들어줌
# 만약 canvas의 크기를 임의로 설정하려면 figure객체를 만들어야함

## Figure객체는 1개 이상의 Axes로 구성되요!
## plot() 함수를 이용하면 자동으로 Axes도 하나 생성해요!
## 코드로 여러개의 Axes를 만들어서(sub plot을 생성해서) 그래프를 그려보아요!
```

```python
plt.style.use('ggplot')

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(1,2,1)  # 1행 2열의 첫번째 axes
ax2 = fig.add_subplot(1,2,2)  # 1행 2열의 두번째 axes

ax1.set_title('서울 -> 경기 인구 이동')
ax1.set_xlabel('연도')
ax1.set_ylabel('이동 인구수')

ax1.plot(s.index,s.values)

ax2.plot(s.index, s.values,
         marker = 'o',  # o는 circle marker, s는 square marker
         markersize=8,
         markerfacecolor='b',
         color='g',   # 선의 색깔
         linewidth=2)

plt.show()
```

```python
# Histogram에 대해서 알아보자
# 변수가 하나인 단변수 데이터의 빈도수를 그래프로 표현한것

# x축을 같은 크기의 여러구간(bin)으로 나누고
# y축은 해당 구간안에 데이터의 개수(빈도, mode)를 표현
# 당연히, x축의 구간(bin)을 조절하면 histogram의 모양은 달라짐

# 실제 데이터를 이용해 그려보자
# Data는 MPG data set 이용

import numpy as np
import pandas as pd

df = pd.read_csv('./data/auto-mpg.csv', header=None) 

df.columns = ['mpg','cylinders','displacement','horsepower',
              'weight','acceleration','year','origin','name']
display(df)

print(df['mpg'].max(), df['mpg'].min())   # mpg column의 최대값과 최솟값을 알아보자  # 46.6 9.0

# matplotlib이 아닌 pandas의 기능을 이용해서 histogram을 그려보자
df['mpg'].plot(kind='hist',
               bins=10, # 최솟값과 최댓값 사이를 10개의 구간으로 분할
               color='blue',
               figsize=(5,5))
```

```python
# Histogram : 단변수에 대한 빈도를 그래프로 표현
# Scatter(산점도) : 서로 다른 두 변수 사이의 관계를 표현
# 2개의 변수의 값을 각각 x축과 y축에 하나씩 놓고 데이터 값이 위치하는 (x,y)좌표를 
# 찾아서 점을 표현

# 코드를 이용해서 scatter를 그려보자
# MPG Data Set 사용

df = pd.read_csv('./data/auto-mpg.csv', header=None) 

df.columns = ['mpg','cylinders','displacement','horsepower',
              'weight','acceleration','year','origin','name']
display(df.head(3))

# 두 개의 변수 사용 -> mpg(연비)와 weight(중량) 이용
# scatter 그려보자
df.plot(kind='scatter',
        x='weight',
        y='mpg',
        color='red',
        s=10,
        figsize=(10,5))
```

```python
# Boxplot
# 단변수의 데이터 분포를 살펴보기 위해 사용
# 이상치를 눈으로 쉽게 파악 가능

import matplotlib.pyplot as plt

df = pd.read_csv('./data/auto-mpg.csv', header=None) 

df.columns = ['mpg','cylinders','displacement','horsepower',
              'weight','acceleration','year','origin','name']
display(df.head(3))


# 제조국 별 연비분포를 boxplot으로 그려보자!
# 제조국(origin) => 1 : USA, 2 : EU, 3 : JPN

plt.boxplot(x=[df.loc[df['origin']==1,'mpg'],
               df.loc[df['origin']==2,'mpg'],
               df.loc[df['origin']==3,'mpg']])
plt.show()
```
