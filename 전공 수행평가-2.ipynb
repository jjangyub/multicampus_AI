```python
import numpy as np
import pandas as pd

df1 = pd.read_csv('./data/weather_daegu.csv', encoding='cp949')
df2 = pd.read_csv('./data/weather_seoul.csv', encoding='cp949')

# 각 데이터에 결측치들 있는 행은 제거함
df1.rename(columns={'평균기온(℃)':'평균기온','최저기온(℃)':'최저기온', '최고기온(℃)':'최고기온'}, inplace=True)
display(df1.head())
print(df1.isnull().sum())
df1.dropna(axis=0, inplace=True)
# df1.info()

df2.rename(columns={'평균기온(℃)':'평균기온','최저기온(℃)':'최저기온', '최고기온(℃)':'최고기온'}, inplace=True)
display(df2.head())

print(df2.isnull().sum())
df2.dropna(axis=0, inplace=True)
df2.info()

display(df1)
display(df2)

### 1. 기상 관측 이래, 서울의 최고 기온이 가장 높았던 날은 언제였고, 몇도인가요?
print('===1번===')
# print(df2.loc[df2['최고기온'].argmax(),['날짜','최고기온']])
max_temp = df2['최고기온'].max()
print(max_temp)

temp_max_date = (df2['최고기온']==max_temp)
display(df2[temp_max_date])


### 2. 역사적으로 일교차가 가장 큰 날짜는 몇년 몇월 몇일 인가요?
daegu_max = (df1['최고기온']-df1['최저기온']).max() 
seoul_max = (df2['최고기온']-df2['최저기온']).max()

print('===2번===')
if daegu_max > seoul_max :
    temp_max_date = ((df1['최고기온']-df1['최저기온'])==daegu_max)
    display(df1[temp_max_date])
    
else :
    temp_max_date = ((df2['최고기온']-df2['최저기온'])==seoul_max)
    display(df2[temp_max_date])
```

```python
### 4. 가장 덥다고 알려진 대구보다 서울이 더 더운날이 가장 많은 연도는 언제인가요?
print('===4번===')
df1_temp = df1['최고기온']
df2_temp = df2['최고기온']
# display(df1_temp)

result_df = pd.merge(df1,df2,on='날짜', how='inner')
display(result_df)

result_df['비교'] = result_df['최고기온_y'] - result_df['최고기온_x']
display(result_df)

year_series = pd.Series()

for y in range(115):
    year = 1909
    count = 0
    year = year + y
    year_search = str(year)
    find_year = result_df[result_df['날짜'].str.contains(year_search)]
    
    for i in range(find_year.shape[0]):
        if (find_year.iloc[i]['비교'] > 0):
            count = count + 1
        
        year_series[year_search] = count

year_series
max_count = year_series.max()
year = (year_series == max_count)

print('4번 답')
print(year_series[year])
```

```python
### 3. 1년 중 평균적으로 일교차가 가장 큰 날짜는 몇월 몇일 인가요?

print('===3번===')

df2['일교차'] = df2['최고기온'] - df2['최저기온']
display(df2)

df2['날짜'] = pd.to_datetime(df2['날짜'],format='%Y-%m-%d')

max_mean = 0

for m in range(12):
    for d in range(31):
        target_month = 1 + m
        target_day = 1 + d
        find = df2.query('날짜.dt.month == @target_month & 날짜.dt.day == @target_day')
        
        if max_mean < find['일교차'].mean() :
            max_mean = find['일교차'].mean()
            max_month = target_month
            max_day = target_day
        
print(max_month, '월', max_day, '일')       
```
