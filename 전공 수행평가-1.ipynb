```python
# 1. 사용자가 평가한 모든 영화의 전체 평균 평점을 출력하자

import numpy as np
import pandas as pd

df = pd.read_csv('./data/ratings.csv')
display(df)

print(df['rating'].mean())
```

```python
# 2. 각 사용자별 평균 평점 구하시오
#    출력시 정렬은 userId로 오름차순 정렬하기

grouped = df.groupby(['userId']).mean()
display(grouped['rating'])              
```

```python
# 3. 각 영화별 평균 평점을 구하시요
#    출력시 정렬은 movieId로 오름차순 정렬
df1 = pd.read_csv('./data/ratings.csv')
# display(df1.head())

df2 = pd.read_csv('./data/movies.csv')
# display(df2.head())

result_df = pd.merge(df1, df2, on='movieId', how='inner')
display(result_df)

grouped =result_df.groupby(['movieId','title']).mean()
display(grouped['rating'])
```

```python
# 4. 평균 평점이 가장 높은 영화의 제목을 출력하시오
#    단, 동률이 있을 경우 모두 출력하고
#    title을 기준으로 오름차순 정렬
df1 = pd.read_csv('./data/ratings.csv')
# display(df)

df2 = pd.read_csv('./data/movies.csv')
# display(df2)

result_df = pd.merge(df1, df2, on='movieId', how='inner')
display(result_df)

grouped =result_df.groupby(['movieId','title']).mean()
grouped_sort = grouped.sort_values(by='rating', ascending=False)
display(grouped_sort)

max_rating = grouped_sort.iloc[0]['rating']
print(max_rating)

# print(grouped_sort['rating']==max_rating)
rating_max = (grouped_sort['rating']==max_rating)

# sort는 복사본 안 만듦
grouped_sort = grouped_sort.sort_values(by='title',ascending=True)
display(grouped_sort[rating_max])
```

```python
# 5. Comedy영화 중 가장 평점이 낮은 영화의 제목을 출력하시오.
#    동률이 있을 경우 모두 출력하고
#    title를 기준으로 오름차순 정렬

result_df = pd.merge(df1, df2, on='movieId', how='inner')
# display(result_df)

comedy_movie = result_df['genres'].str.contains('Comedy')
# display(comedy_movie)

grouped_comedy = result_df[result_df['genres'].str.contains('Comedy')]
# display(grouped_comedy)

grouped =grouped_comedy.groupby(['movieId','title']).mean()
display(grouped)

grouped_sort = grouped.sort_values(by='rating', ascending=True)
display(grouped_sort)

min_rating = grouped_sort.iloc[0]['rating']
print(min_rating)

rating_min = (grouped_sort['rating']==min_rating)
grouped_sort = grouped_sort.sort_values(by='title',ascending=True)
display(grouped_sort[rating_min])


# genres_df = result_df['genres'].str.get_dummies(sep='|')
# genres_df.head()

# result_df = result_df.drop(['genres'],axis=1)
# result_df = result_df.join((genres_df))
# display(result_df)
```

```python
# 6. 2015년도에 평가된 모든 Romance 영화의 평균평점은?
from datetime import datetime
import time

result_df = pd.merge(df1, df2, on='movieId', how='inner')
# display(result_df)

date = []

for i in range(100836):
    t = result_df.iloc[i]['timestamp']
    dt = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(t))
    date.append(dt)
result_df['datetime']= date
display(result_df)
```

```python
# 7. 모든 영화장르 중 사용자 평점이 가장 좋은 영화장르는 무엇인가요?
#    동률이 있으면 영화장르를 기준으로 오름차순 정렬

result_df = pd.merge(df1, df2, on='movieId', how='inner')
# display(result_df)

df_sort = result_df.sort_values(by='rating', ascending=False)
# display(df_sort)

max_rating = df_sort.iloc[0]['rating']
# print(max_rating)

rating_max = (df_sort['rating']==max_rating)

best = df_sort.sort_values(by='genres',ascending=True)
display(best[rating_max])
```
