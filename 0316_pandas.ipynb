```python
import numpy as np
import pandas as pd

# ndarray를 생성해봐요!
arr = np.array([1, 2, 3, 4, 5],dtype=np.float64)
print(arr)

# Serires를 생성해봐요!
s = pd.Series([1, 2, 3, 4, 5],dtype=np.float64)
print(s)   # index와 value가 같이 출력
# 0    1.0
# 1    2.0
# 2    3.0
# 3    4.0
# 4    5.0
dtype: float64
print(s.values)   # [1. 2. 3. 4. 5.] ndarray
print(s.index)   # RangeIndex(start=0, stop=5, step=1)  => pandas의 객체
print(s.dtype)   # float64 (실수)
```

```python
s = pd.Series([1, 2, 3, 4, 5], 
              dtype=np.float64,
              index=['c','b','a','kk','홍길동'])
print(s)

# 만약 인덱스를 지정해서 serires를 만들면 숫자 index는 어떻게 되나요?
# 숫자 index는 그대로 사용가능. 내부에 숨어있다!
print(s[0])
print(s['c'])

# print(s[100])   # Error
```

```python
s = pd.Series([1, 2, 3, 4, 5], 
              dtype=np.float64,
              index=[0, 2, 100, 6, 9])
print(s)
print(s[0])   # 1.0
# print(s[1])   # Error # 지정 index가 숫자면 숨어있던 index는 날아간다!
print(s[100])

# 지정 index에 같은 index가 존재하면 어떻게 되나요?
# 가능해요!
s = pd.Series([1, 2, 3, 4, 5], 
              dtype=np.float64,
              index=['c','b','c','k','m'])
print(s['c'])   # c    1.0
                # c    3.0
                # dtype: float64
```

```python
# indexing, slicing
# indexing은 ndarray나 list와 비슷함

s = pd.Series([1, 2, 3, 4, 5], 
              dtype=np.float64,
              index=['c','b','a','k','m'])
print(s)
print(s[0:3])   # 숫자 index를 이용한 slicing(ndarray와 list와 동일)
print(s['c':'a'])   # 지정 index를 사용한 slicing 

```

```python
# boolean indexing, fancy indexing 역시 적용가능
s = pd.Series([1, 2, 3, 4, 5], 
              dtype=np.float64,
              index=['c','b','a','k','m'])
print(s)

# 짝수만 뽑아내는 boolean indexing
print(s[s % 2 == 0])

# fancy indexing
print(s[[0,2]])
```

```python
# series를 만드는 또 다른 방법
# index를 key로, 저장하는 값을 value로 구성되는 자료구조
# s = pd.Series([1, 2, 3, 4, 5],
#               dtype=np.float64,
#               index=['c','b','a','k','m'])

# dictionary를 이용해서 series를 생성
my_dict = {'서울' : 100,
           '인천' : 500,
           '제주' : 200
}
s = pd.Series(my_dict)
print(s)
```

```python
# DataFrame에 대해 알아보아요
# 그림으로 Series의 집합이 DataFrame이라는걸 알아봤어요!

# DataFrame은 dictionary를 이용해서 만들어요!

my_dict = {'이름' : ['홍길동', '아이유', '김연아', '신사임당'],
           '학년' : [4, 3, 1, 2],
           '학점' : [1.5, 2.4, 3.9, 3.2]}

# DataFrame의 기본 속성
df = pd.DataFrame(my_dict)
display(df)
print(df.shape)   # (4, 3)
print(df.values)   # 2차원 ndarray
print(df.size)   # DataFrame안의 요소 개수 => 12
print(df.ndim)   # 2차원
print(df.index)   # 행 index - RangeIndex(start=0, stop=4, step=1)
print(df.columns)   # 열 index - Index(['이름', '학년', '학점'], dtype='object')

# DataFrame의 column명과 index명을 변경
# rename()

my_dict = {'이름' : ['홍길동', '아이유', '김연아', '신사임당'],
           '학년' : [4, 3, 1, 2],
           '학점' : [1.5, 2.4, 3.9, 3.2]}
df = pd.DataFrame(my_dict)
display(df)

# inplace 옵션의 값이 True이면 원본을 수정하고 복사본을 만들지 않음!
# inplace 옵션의 값이 False면 원본을 그냥두고 복사본을 만들었어요!

new_df = df.rename(columns = {'이름' : '학생이름',
                              '학점' : '평균평점'},
                  index={0:'one'},
                  inplace=False)
display(new_df)
```

```python
# DataFrame의 특정 column을 index로 설정
# 예를 들어서 이름 column을 index로 설정해 보아요!
# set_index()

import numpy as np
import pandas as pd

my_dict = {'이름': ['홍길동', '아이유', '김연아', '신사임당'],
           '학년': [4, 3, 1, 2],
           '학점': [1.5, 2.4, 3.9, 3.2]}

df = pd.DataFrame(my_dict)
display(df)

new_df = df.set_index('이름', 
                      inplace=False)
display(new_df)
```
